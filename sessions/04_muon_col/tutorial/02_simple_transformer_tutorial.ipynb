# Pixel Transformer tutorial: Improving the Model

## First, a few words on the goal of this tutorial

With the BDT, we classified pixel clusters using engineered cluster quantities. With a transformer we want to classify the same objects using the raw pixel hits inside each cluster. These raw hits contain `(E, t, x, y)` values and all of the cluster quantities are derived from them. While there are only 2 spatial coordinates, all of the hits are on a single layer of the detector so, naively, this should be enough.

Our goal in this notebook is to see if the Transformer can beat the BDT.

You can take one of the routes below to improve them model or you can design your own!

## Outline

1. Load the raw-hit pixel-cluster data.
2. Build a basic Transformer from PyTorch layers.
3. Choose an improvement to the model.
4. Load the pretrained weights, or train the model if you chose `custom`.
5. Compare against the BDT reference.

## The Base Transformer

The base model mirrors what we saw during the earlier lecture:

1. Project each hit into a hidden representation.
2. Let hits exchange information through self-attention.
3. Mask padded hit slots so they do not contribute.
4. Pool the learned hit representations into one cluster representation.
5. Classify the cluster as BIB or signal.


```python
class Attention(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.q = nn.Linear(dim, dim)
        self.k = nn.Linear(dim, dim)
        self.v = nn.Linear(dim, dim)
        self.scale = dim ** -0.5

    def forward(self, x, mask):
        # TODO: Implement the attention mechanism using the query, key, and value projections.
        q = ...
        k = ...
        v = ...

        attn = (q @ k.transpose(-2, -1)) * self.scale
        attn = attn.softmax(dim=-1)
        x = attn @ v

        return x * mask


class TransformerBlock(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.att = Attention(dim)
        self.proj1 = nn.Linear(dim, dim)
        self.proj2 = nn.Linear(dim, dim)
        self.activation = nn.GELU()
        self.norm1 = nn.LayerNorm(dim)
        self.norm2 = nn.LayerNorm(dim)

    def forward(self, x, mask):
        x = x + self.att(self.norm1(x), mask)
        x = self.activation(self.proj1(x)) * mask
        x = x + self.proj2(self.norm2(x)) * mask
        return x


class SimplePixelTransformer(nn.Module):
    def __init__(self, input_dim=4, hidden_dim=64, num_layers=3, num_classes=2):
        super().__init__()
        self.input_layer = nn.Linear(input_dim, hidden_dim)
        self.blocks = nn.ModuleList([...])  # TODO: Create a list of TransformerBlock instances of the proper length.
        self.output_layer = nn.Sequential(
            nn.LayerNorm(hidden_dim),
            nn.Linear(hidden_dim, hidden_dim),
            nn.GELU(),
            nn.Linear(hidden_dim, num_classes),
        )

    def forward(self, x):
        # A hit is real if the energy feature is nonzero.
        mask = (x[:, :, 0] != 0).unsqueeze(-1).float()

        # TODO: Pass x through the input layer then multiply by the mask to zero out padded hits.
        x = ...
        for block in self.blocks:
            x = block(x, mask)

        # Sum pooling keeps some occupancy information; the sqrt keeps the scale moderate.
        x = (x * mask).sum(dim=1) / (mask.shape[1] ** 0.5)
        return self.output_layer(x)
```

## Route 2: Pooling

The motivation here is that we know the BDT puts higher importance on features like `cluster_size`, which is a difference in extrema. Providing sum & mean pooling might allow more access to these.

An idea for a custom model might be one where you directly calculate the extrema of the transformer outputs and pass them to the MLP along with some form of pooling.

## Route 3: Dropout

Dropout randomly removes small pieces of information during training. At evaluation time it is disabled, but during training the model experiences that extra noise, which can make the model less brittle.

## Route 5: Primitive Detector Context

The raw local pixel coordinates do not fully describe where the cluster lives in the detector. Although this shouldn't matter since all the hits lie on one layer, it is possible that our model cannot figure this out. Add a few cluster-level quantities: `cluster_x`, `cluster_y`, `cluster_z`, `cluster_r`, and `cluster_size_x`.

The context arrays use the same train/validation/test indices as the raw-hit loaders.


```python
CONTEXT_KEYS = ['cluster_x', 'cluster_y', 'cluster_z', 'cluster_r','cluster_size_x']
_CONTEXT_LOADERS = None


def get_context_loaders():
    global _CONTEXT_LOADERS
    if _CONTEXT_LOADERS is None:
        _CONTEXT_LOADERS = make_context_loaders(
            SIGNAL_H5,
            BIB_H5,
            CONTEXT_KEYS,
            train_loader,
            val_loader,
            test_loader,
            idx_train,
            idx_val,
            idx_test,
            BATCH_SIZE,
        )
    return _CONTEXT_LOADERS
```
